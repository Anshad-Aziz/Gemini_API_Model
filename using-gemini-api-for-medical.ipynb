```python
import google.generativeai as genai
import os
```

```python
from IPython.display import Markdown
model=genai.GenerativeModel('gemini-pro')
```

```python
response=model.generate_content("You are a doctor! Give me Just a single health tip!")
Markdown(response.text)
```

```python
age="20"
gender="male"
occupation="student"
nationality="Indian"
weight="70"
health_insurance="yes"

existing_medical_symptoms="cough& cold,runny nose,frequent fever"
chronic_diseases="none"
allergies="none"
previous_surgeries="none"
current_medications="antibiotic medicine for cough & cold"

smoking_status="none"
alcohol_intake="none"
occasional_drug_use="none"
```

```python
# creating a prompt template for our patient
prompt = "Patient is a {0} year old {1}".format(age, gender)
prompt += "Patient is {0} and is works as a {1}.".format(nationality, occupation)
prompt += "Patient weighs {0} and their health insurance status is {1}".format(weight,health_insurance)
prompt += "Patient also has some existing medical symptoms like {0}.".format(existing_medical_symptoms)
prompt += "Patient's chronic disease status is {0}, allergy status is {1}.".format(chronic_diseases, allergies)
prompt += "Patient's previous surgeries are {0} and current medications are {1}.".format(previous_surgeries, current_medications)
prompt += "Patient's smoking status is {0}, alcohol intake is {1} and occasional drug use is {2}.".format(smoking_status, alcohol_intake, occasional_drug_use)
prompt += "You are a renowned doctor. You know everything about diseases and their cures. You will help the patient by recommending him/her medicines and steps he/she should take to get cured at the earliest using the symptoms provided above."
```

```python
response=model.generate_content(prompt)
Markdown(response.text)
```
